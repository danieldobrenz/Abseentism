## Creating a checkpoint

```python
df_checkpoint_1 = df.copy()
```

## Changing the 'Date' column to a format that is Python readable

```python
df_checkpoint_1['Date'] = (df_checkpoint_1['Date']) = pd.to_datetime(df_checkpoint_1['Date'], format= '%d/%m/%Y')
```

```python
df_checkpoint_1['Date'].head()
```

## Extracting the 'month' from the date column

```python
df_checkpoint_1['Date'][0]
```

```python
df_checkpoint_1['Date'][0].month
```

```python
# here we will create a list that has just the values of the 'month' value 
# then we will add the list to a new column to the data frame
list_months = []
list_months
```

```python
df_checkpoint_1.shape
```

```python
# here we create a for loop that iterates through the dates, then extracts the month values
for i in range(700):
    list_months.append(df_checkpoint_1['Date'][i].month)
```

```python
df_checkpoint_1['Month Value'] = list_months
```

```python
df_checkpoint_1.head
```

## Extract the Day of the Week Values:

```python
df_checkpoint_1['Date'][699].weekday()
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
df_checkpoint_1['Day of the Week'] = df_checkpoint_1['Date'].apply(date_to_weekday)
```

```python
df_checkpoint_1.head()
```

```python
df_checkpoint_1 = df_checkpoint_1.drop(['Date'], axis= 1)
```
